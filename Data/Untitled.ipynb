```python
import pandas as pd

# Step 1: Count unique scans per patient in master_dexa
scan_counts = (
    master_dexa.groupby('Patient Name')['Unique ID']
    .nunique()
    .reset_index(name='scan_count')
)

# Step 2: Identify patients with 5 or more scans
patients_5plus = scan_counts[scan_counts['scan_count'] >= 5]['Patient Name']

# Step 3: Filter both datasets
master_dexa_filtered = master_dexa[master_dexa['Patient Name'].isin(patients_5plus)]
composition_filtered = composition[composition['Patient Name'].isin(patients_5plus)]

# Step 4: (optional) Check results
print(f"Number of patients with ≥5 scans: {patients_5plus.nunique()}")
print(f"master_dexa_filtered shape: {master_dexa_filtered.shape}")
print(f"composition_filtered shape: {composition_filtered.shape}")
```

```python
import pandas as pd
import os

# Paths to your CSV files - adjust these to match your actual paths
working_csv = "master_dexa_data_Nov.csv"  # REPLACE THIS
broken_csv = "master_dexa_data.csv"       # REPLACE THIS

print("=" * 80)
print("CSV COMPARISON TOOL")
print("=" * 80)

# Check if both files exist
print("\n1. Checking files...")
if not os.path.exists(working_csv):
    print(f"   ✗ Working CSV not found: {working_csv}")
    print("   Please update the working_csv path in this script")
    exit()
else:
    print(f"   ✓ Working CSV found: {working_csv}")
    print(f"     Size: {os.path.getsize(working_csv):,} bytes")

if not os.path.exists(broken_csv):
    print(f"   ✗ Broken CSV not found: {broken_csv}")
    print("   Please update the broken_csv path in this script")
    exit()
else:
    print(f"   ✓ Broken CSV found: {broken_csv}")
    print(f"     Size: {os.path.getsize(broken_csv):,} bytes")

# Load both CSVs
print("\n2. Loading CSVs...")
try:
    df_working = pd.read_csv(working_csv)
    print(f"   ✓ Working CSV loaded: {df_working.shape[0]} rows, {df_working.shape[1]} columns")
except Exception as e:
    print(f"   ✗ Failed to load working CSV: {e}")
    exit()

try:
    df_broken = pd.read_csv(broken_csv)
    print(f"   ✓ Broken CSV loaded: {df_broken.shape[0]} rows, {df_broken.shape[1]} columns")
except Exception as e:
    print(f"   ✗ Failed to load broken CSV: {e}")
    exit()

# Compare basic structure
print("\n3. Basic Structure Comparison:")
print(f"   Working: {df_working.shape[0]} rows × {df_working.shape[1]} columns")
print(f"   Broken:  {df_broken.shape[0]} rows × {df_broken.shape[1]} columns")

if df_working.shape[0] != df_broken.shape[0]:
    print(f"   ⚠ Row count differs by {abs(df_working.shape[0] - df_broken.shape[0])} rows")
if df_working.shape[1] != df_broken.shape[1]:
    print(f"   ⚠ Column count differs by {abs(df_working.shape[1] - df_broken.shape[1])} columns")

# Compare columns
print("\n4. Column Comparison:")
working_cols = set(df_working.columns)
broken_cols = set(df_broken.columns)

if working_cols == broken_cols:
    print(f"   ✓ Both have identical {len(working_cols)} columns")
    print(f"   Columns: {list(df_working.columns)}")
else:
    print(f"   ✗ Columns are different!")
    
    only_in_working = working_cols - broken_cols
    only_in_broken = broken_cols - working_cols
    
    if only_in_working:
        print(f"\n   Columns only in WORKING CSV ({len(only_in_working)}):")
        for col in sorted(only_in_working):
            print(f"      - '{col}'")
    
    if only_in_broken:
        print(f"\n   Columns only in BROKEN CSV ({len(only_in_broken)}):")
        for col in sorted(only_in_broken):
            print(f"      - '{col}'")
    
    common_cols = working_cols & broken_cols
    if common_cols:
        print(f"\n   Common columns ({len(common_cols)}):")
        for col in sorted(common_cols):
            print(f"      - '{col}'")

# Check for Patient Name column specifically
print("\n5. 'Patient Name' Column Check:")
if 'Patient Name' in df_working.columns:
    print(f"   ✓ Working CSV has 'Patient Name'")
    print(f"     - {df_working['Patient Name'].notna().sum()} non-null values")
    print(f"     - {df_working['Patient Name'].nunique()} unique patients")
else:
    print(f"   ✗ Working CSV missing 'Patient Name'")

if 'Patient Name' in df_broken.columns:
    print(f"   ✓ Broken CSV has 'Patient Name'")
    print(f"     - {df_broken['Patient Name'].notna().sum()} non-null values")
    print(f"     - {df_broken['Patient Name'].nunique()} unique patients")
else:
    print(f"   ✗ Broken CSV missing 'Patient Name'")

# Compare Scan Date column as suspected
print("\n6. 'Scan Date' Column Analysis:")
if 'Scan Date' in df_working.columns and 'Scan Date' in df_broken.columns:
    print("   Working CSV - Scan Date:")
    print(f"     - dtype: {df_working['Scan Date'].dtype}")
    print(f"     - Non-null: {df_working['Scan Date'].notna().sum()}/{len(df_working)}")
    print(f"     - Sample values: {df_working['Scan Date'].head(3).tolist()}")
    
    print("   Broken CSV - Scan Date:")
    print(f"     - dtype: {df_broken['Scan Date'].dtype}")
    print(f"     - Non-null: {df_broken['Scan Date'].notna().sum()}/{len(df_broken)}")
    print(f"     - Sample values: {df_broken['Scan Date'].head(3).tolist()}")
    
    # Check if there are problematic date values
    try:
        working_dates = pd.to_datetime(df_working['Scan Date'], errors='coerce')
        broken_dates = pd.to_datetime(df_broken['Scan Date'], errors='coerce')
        
        print("\n   Date parsing check:")
        print(f"     - Working: {working_dates.isna().sum()} unparseable dates")
        print(f"     - Broken: {broken_dates.isna().sum()} unparseable dates")
        
        if broken_dates.isna().sum() > 0:
            print("\n   ⚠ Problematic dates in broken CSV:")
            bad_dates = df_broken[broken_dates.isna()]['Scan Date'].unique()
            for date in bad_dates[:5]:  # Show first 5
                print(f"      - '{date}'")
    except Exception as e:
        print(f"   Error parsing dates: {e}")

# Data type comparison for common columns
if working_cols == broken_cols:
    print("\n7. Data Type Comparison:")
    dtype_diffs = []
    for col in df_working.columns:
        if df_working[col].dtype != df_broken[col].dtype:
            dtype_diffs.append(col)
            print(f"   ⚠ '{col}': {df_working[col].dtype} (working) vs {df_broken[col].dtype} (broken)")
    
    if not dtype_diffs:
        print("   ✓ All data types match")

# First few rows comparison
print("\n8. First 3 Rows - Working CSV:")
print(df_working.head(3))

print("\n9. First 3 Rows - Broken CSV:")
print(df_broken.head(3))

# Check for empty rows
print("\n10. Empty Row Check:")
working_empty = df_working.isna().all(axis=1).sum()
broken_empty = df_broken.isna().all(axis=1).sum()
print(f"   Working: {working_empty} completely empty rows")
print(f"   Broken: {broken_empty} completely empty rows")

# Null value comparison
print("\n11. Null Values Summary:")
print("\n   Working CSV:")
working_nulls = df_working.isna().sum()
if working_nulls.sum() > 0:
    print(working_nulls[working_nulls > 0])
else:
    print("   No null values")

print("\n   Broken CSV:")
broken_nulls = df_broken.isna().sum()
if broken_nulls.sum() > 0:
    print(broken_nulls[broken_nulls > 0])
else:
    print("   No null values")

print("\n" + "=" * 80)
print("COMPARISON COMPLETE")
print("=" * 80)

# Summary recommendation
print("\n📋 SUMMARY & RECOMMENDATIONS:")
if df_working.shape[1] == 0 or df_broken.shape[1] == 0:
    print("   🚨 One or both files have NO COLUMNS - likely a parsing issue!")
    print("   Try opening the CSV in a text editor to check the format")
elif working_cols != broken_cols:
    print("   🚨 Column mismatch detected - this is likely your issue!")
    print("   The broken CSV is missing critical columns")
elif df_working.shape != df_broken.shape:
    print("   ⚠ Different number of rows - data may be incomplete")
else:
    print("   ℹ Files have same structure - issue may be in data values")
    print("   Check the Scan Date analysis above for problematic values")
```
